# HOUSING PRICE (MULTIVARIABLE)

## IMPORT NECESSARY MODULES

```python
import math as m
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## DATASET

```python
dt=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/2_linear_reg_multivariate/homeprices.csv')
Y=dt.price
X=dt.drop(['price'],axis=1)
X
```

```python
median=m.floor(X.bedrooms.median())
X.bedrooms=X.bedrooms.fillna(median)
X
```

## FITTING

```python
lr=LinearRegression()
lr.fit(X,Y)
```

## COEFFICIENTS AND INTERCEPT

```python
print("coefficient:",lr.coef_)
print("intercept:",lr.intercept_)
```

## PREDICTING THE VALUES (given)
 i) 3000 sqft area 3 bedrooms, 40 year old
 ii) 2500 sqft area 4 bedrooms, 5 year old

```python
test=pd.DataFrame([[3000,3,40],[2500,4,5]],columns=["area","bedrooms","age"])
test
```

```python
print(f"Test Case 1 : {lr.predict(test)[0]} \nTest Case 2 : {lr.predict(test)[1]}")
```
