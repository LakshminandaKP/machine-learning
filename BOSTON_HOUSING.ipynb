# BOSTON HOUSING

## IMPORT MODULES

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np
```

## DATASET

```python
dt=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/BostonHousing.csv")
dt
```

## SEPARATE X AND Y

```python
Y=dt['medv']
X=dt.drop('medv',axis=1)
X
```

## SPLITTING

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=50,test_size=0.2)
X_train
Y_train
```

## FITTING

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

## PREDICTING

```python
Y_pred=lr.predict(X_test)
Y_pred
```

## EVALUVATION

```python
mse=mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
r2,mse
```

```python
result=pd.DataFrame(["LINEAR REGRESSION",mse,r2]).transpose()
result.columns=['method','mean squared error','r squared']
result
```

## PLOTTING

```python
plt.figure(figsize=[5,5])
plt.scatter(Y_test,Y_pred,color='blue',label='datapoints',alpha=0.2)
pol=np.polyfit(Y_test,Y_pred,1)
p=np.poly1d(pol)

plt.plot(Y_test,p(Y_test),c='#111111',label='regression line')
plt.title("BOSTON HOUSE LR")
plt.xlabel('Y test')
plt.ylabel('Y pred')
plt.legend()
plt.show()
```

# RANDOM FOREST

## NECESSARY MODULES

```python
from sklearn.ensemble import RandomForestRegressor
```

## FITTING

```python
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train,Y_train)
```

## PREDICTING

```python
rf_y_pred=rf.predict(X_test)
```

## EVALUATION

```python
rf_mse=mean_squared_error(Y_test,rf_y_pred)
rf_r2=r2_score(Y_test,rf_y_pred)
```

```python
rf_result=pd.DataFrame(["RANDOM FOREST",rf_mse,rf_r2]).transpose()
rf_result.columns=['method','mean squared error','r squared']
rf_result
```

## PLOTTING

```python
plt.figure(figsize=[5,5])
plt.scatter(Y_test,rf_y_pred,color='blue',alpha=0.2,label='data point')
pol=np.polyfit(Y_test,rf_y_pred,1)
p=np.poly1d(pol)
plt.plot(Y_test,p(Y_test),color='red',label='regression line')
plt.title('BOSTON HOUSING RF')
plt.xlabel('Ytest')
plt.ylabel('Y predict')
plt.legend()
plt.show()
```

## MAKE BOTH LR AND RF IN A SINGLE TABLE

```python
all_results=pd.concat([result,rf_result]).reset_index(drop=True)
all_results
```
