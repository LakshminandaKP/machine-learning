# HOUSING PRICES

## IMPORT MODULE

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


```

##DATA SET


```python
dt=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/1_linear_reg/homeprices.csv')
dt
Y=dt.price
X=dt.drop("price",axis=1)
X
```

## FITTING

```python
lr=LinearRegression()
m=lr.fit(X,Y)
m
```

## PREDICTING

```python
y_pred=m.predict(X)
```

## PLOTTING

```python
plt.figure(figsize=[5,5])
plt.scatter(X,Y,color='blue',label='datapoints',marker='+')
plt.title('HOUSING PRICE')
plt.xlabel('area')
plt.ylabel('price')
plt.plot(X,y_pred,color='red',label='regression line')
plt.legend()
plt.show()
```

```python
pred1=m.predict([[3300]])
pred2=m.predict([[56000]])
print(f'pred1={pred1}, pred2={pred2}')
```
